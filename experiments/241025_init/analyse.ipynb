```python
import plot_utils
from plot_utils import Roller,LogFile,Fig,LogFileType
import rolling_funcs as fn
```

```python

window_len_secs = 20.0
log_files = {
    "shrew attack": LogFile(log_type=LogFileType.GOT, path= "./logs/mac/2024-10-25_14-40-50.238948.txt"),
    "service" : LogFile(log_type=LogFileType.GOT, path="./logs/kleene/2024-10-25_14-25-37.472506.txt"),
    "service under attack" : LogFile(log_type=LogFileType.GOT, path="./logs/kleene/2024-10-25_14-38-56.215879.txt"),
}
rollers = [
    Roller("OK rate", fn.count_ok, rate=True),
    Roller("ERR rate", fn.count_err, rate=True)
]

fig,ax = plot_utils.fig([Fig('10 RPS following a systemd restart, then reboot and a shrew attack.', 'time (s)', 'events per second')])
plot_utils.overlay_rolling(ax,log_files,rollers,window_secs=window_len_secs, const_stride_secs=0.5, zeroed_times=False)
ax.legend()
fig.show()
```
